```python
import pandas as pd
import uuid
```

```python
# read datasets for US, Canada, Japan, India, and Brazil
df_br = pd.read_csv('datasets\BR_youtube_trending_data.csv')
df_ca = pd.read_csv('datasets\CA_youtube_trending_data.csv')
df_in = pd.read_csv('datasets\IN_youtube_trending_data.csv')
df_jp = pd.read_csv('datasets\JP_youtube_trending_data.csv')
df_us = pd.read_csv('datasets\\US_youtube_trending_data.csv')
```

```python
# add country code and information
df_br['countryName'] = 'brazil'
df_br['countryCode'] = 1

df_ca['countryName'] = 'canada'
df_ca['countryCode'] = 2

df_in['countryName'] = 'india'
df_in['countryCode'] = 3

df_jp['countryName'] = 'japan'
df_jp['countryCode'] = 4

df_us['countryName'] = 'us'
df_us['countryCode'] = 5
```

```python
df_br = df_br.tail(4000)
df_ca = df_ca.tail(4000)
df_in = df_in.tail(4000)
df_jp = df_jp.tail(4000)
df_us = df_us.tail(4000)
```

```python
df_merged = pd.concat([df_br, df_ca, df_in, df_jp, df_us],ignore_index=True)
```

```python
df_merged.head(5)
```

```python
# channel table
df_channel = df_merged[['channelId', 'channelTitle']]
df_channel['channelTitle'] = df_channel['channelTitle'].str.lower()
df_channel = df_channel.drop_duplicates(subset=['channelId'])
df_channel
```

```python
# category table
df_category = df_merged[['video_id', 'categoryId', 'tags']]
# category_key = [str(uuid.uuid4()) for i in range(df_category.shape[0])]
# df_category['category_key'] = category_key
# df_category = df_category[['category_key', 'video_id', 'categoryId', 'tags']]
df_category = df_category.drop_duplicates(subset=['video_id'])
df_category['tags'] = df_category['tags'].str.lower()
df_category
```

```python
# video table
df_video = df_merged[['video_id', 'channelId', 'publishedAt', 'title', 'thumbnail_link', 'description', 'countryName']]
df_video['publishedAt'] = df_video['publishedAt'].str[:13]
df_video = df_video.drop_duplicates(subset=['video_id'])
df_video
```

```python
# table for stats at published date

df_video_stats = df_merged[['trending_date',  'video_id', 'view_count', 'likes', 'dislikes', 'comment_count', 'comments_disabled', 'ratings_disabled']]
#df_video_stats['times_trending'] = df_video_stats.groupby('video_id')['video_id'].transform('count')
df_video_stats = df_video_stats.drop_duplicates(subset=['trending_date', 'video_id'])
df_video_stats['comments_disabled'] = df_video_stats['comments_disabled'].astype(int)
df_video_stats['ratings_disabled'] = df_video_stats['ratings_disabled'].astype(int)
df_video_stats
```

```python
df_country = df_merged[['countryName', 'countryCode']].drop_duplicates()
df_country
```

```python
# drop tables
drop = '''
DROP TABLE IF EXISTS `manages`;
DROP TABLE IF EXISTS `user`; 
DROP TABLE IF EXISTS `country`;
DROP TABLE IF EXISTS `video_stats`;
DROP TABLE IF EXISTS `category`;
DROP TABLE IF EXISTS `video`;
DROP TABLE IF EXISTS `channel`;
'''
```

```python
# table for country
country = '''
/* Table structure for table country */
CREATE TABLE `country` (
    `country_code` INT(11) PRIMARY KEY NOT NULL,
    `country` VARCHAR(20) NOT NULL
);

'''

country_inserts = '''
/* Data for `country` */
INSERT INTO `country`(`country_code`, `country`)
VALUES
(0, 'Brazil'),
(1, 'Canada'),
(2, 'India'),
(3, 'Japan'),
(4, 'United States');
'''
```

```python
# create table for users
user = '''
/* Table structure for table 'user' */
CREATE TABLE `user` (
    `user_id` INT(11) PRIMARY KEY NOT NULL,
    `country_code` INT(11) NOT NULL,
    `username` VARCHAR(50) NOT NULL,
    `password` VARCHAR(50) NOT NULL,
    `email_id` VARCHAR(50) NOT NULL,
    FOREIGN KEY(`country_code`) REFERENCES `country`(`country_code`) ON UPDATE CASCADE ON DELETE CASCADE
);
'''
```

```python
# create table for channel
channel = '''
/* Table structure for 'channel' */
CREATE TABLE `channel` (
    `channel_id` VARCHAR(50) PRIMARY KEY NOT NULL,
    `channel_title` VARCHAR(255) NOT NULL
);
'''
channel_inserts = '''
/*Data for 'channel'*/
INSERT INTO `channel`(`channel_id`, `channel_title`)
VALUES
'''
count = 0
for i, row in df_channel.iterrows():
    channel_title = row["channelTitle"].replace("'", "''")
    if count == df_channel.shape[0] - 1:
        channel_inserts+=(f'(\'{row["channelId"]}\', \'{channel_title}\');')
        break
    channel_inserts+=(f'(\'{row["channelId"]}\', \'{channel_title}\'),\n')
    count+=1
```

```python
# create table for video
video = ''' 

/* Table structure for 'video' */
CREATE TABLE `video` (
    `video_id` VARCHAR(50) PRIMARY KEY NOT NULL,
    `channel_id` VARCHAR(50) NOT NULL,
    `publishedAt` VARCHAR(50) NOT NULL,
    `title` TEXT NOT NULL,
    `thumbnail_link` VARCHAR(255) NOT NULL,
    `description` TEXT,
    `country` VARCHAR(50) NOT NULL,
    FOREIGN KEY(`channel_id`) REFERENCES `channel`(`channel_id`) ON UPDATE CASCADE ON DELETE CASCADE
);
'''

video_inserts = '''
/*Data for 'video'*/
INSERT INTO `video`(`video_id`,`channel_id`,`publishedAt`,`title`,`thumbnail_link`,`description`, `country`)
VALUES
'''
count = 0
df_video['description'] = df_video['description'].astype(str)
for i, row in df_video.iterrows():
    video_id = row['video_id']
    channelId = row['channelId']
    publishedAt = row['publishedAt']
    title = row["title"].replace("'", "''")
    thumbnail = row['thumbnail_link']
    description = row['description'].replace("'", "''")
    countryName = row['countryName']
    # if count == df_video.shape[0] - 1:
    if count == df_video.shape[0]-1:
        video_inserts+=(f'(\'{video_id}\', \'{channelId}\', \'{publishedAt}\', \'{title}\', \'{thumbnail}\', \'{description}\', \'{countryName}\');')
        break
    count+=1
    video_inserts+=(f'(\'{video_id}\', \'{channelId}\', \'{publishedAt}\', \'{title}\', \'{thumbnail}\', \'{description}\', \'{countryName}\'), \n')
#     if count == 2000:
#         break
# video_inserts = video_inserts[0:len(video_inserts)-1] + ';'
```

```python
# create table for video stats
video_stats = ''' 

/* Table structure for 'video_stats' */
CREATE TABLE `video_stats` (
    `trending_date` VARCHAR(50) NOT NULL,
    `video_id` VARCHAR(50) NOT NULL,
    `view_count` INT(11), 
    `likes` INT(11),
    `dislikes` INT(11),
    `comment_count` INT(11),
    `comments_disabled` INT(11),
    `ratings_disabled` INT(11),
    PRIMARY KEY (`trending_date`, `video_id`),
    FOREIGN KEY (`video_id`) REFERENCES `video` (`video_id`) ON UPDATE CASCADE ON DELETE CASCADE
);
'''

video_stats_inserts = '''
/*Data for 'video_stats'*/
INSERT INTO `video_stats`(`trending_date`, `video_id`, `view_count`, `likes`, `dislikes`, `comment_count`, `comments_disabled`, `ratings_disabled`)
VALUES
'''
count = 0
for i, row in df_video_stats.iterrows():
    trending_date = row['trending_date']
    video_id = row['video_id']
    view_count = row['view_count']
    likes = row["likes"]
    dislikes = row['dislikes']
    comment_count = row['comment_count']
    comments_disabled = row['comments_disabled']
    ratings_disabled = row['ratings_disabled']
    # times_trending = row['times_trending']

    if count == df_video_stats.shape[0] - 1:
        video_stats_inserts+=(f'(\'{trending_date}\', \'{video_id}\', \'{view_count}\', \'{likes}\', \'{dislikes}\', \'{comment_count}\', \'{comments_disabled}\', \'{ratings_disabled}\');')
        continue
    count+=1
    video_stats_inserts+=(f'(\'{trending_date}\', \'{video_id}\', \'{view_count}\', \'{likes}\', \'{dislikes}\', \'{comment_count}\', \'{comments_disabled}\', \'{ratings_disabled}\'), \n')
print(count)
```

```python
# create table for category
category = ''' 

/* Table structure for 'category' */
CREATE TABLE `category` (
    `categoryId` INT(11) NOT NULL,
    `video_id` VARCHAR(50) NOT NULL,
    `tags` TEXT,
    PRIMARY KEY (`categoryId`, `video_id`),
    FOREIGN KEY (`video_id`) REFERENCES `video` (`video_id`) ON UPDATE CASCADE ON DELETE CASCADE
);
'''

category_inserts = '''
/*Data for 'category'*/
INSERT INTO `category`(`categoryId`, `video_id`, `tags`)
VALUES
'''
count = 0
for i, row in df_category.iterrows():
    categoryId = row['categoryId']
    video_id = row['video_id']
    tags = row['tags'].replace("'", "''")

    if count == df_category.shape[0] - 1:
        category_inserts+=(f'(\'{categoryId}\', \'{video_id}\', \'{tags}\');')
        continue
    count+=1
    category_inserts+=(f'(\'{categoryId}\', \'{video_id}\', \'{tags}\'), \n')
```

```python
# table for relationship manages
manages  = ''' 
/* Table for the relationship manages between user and channel */
CREATE TABLE `manages` (
    `user_id` INT(11) NOT NULL,
    `channel_id` VARCHAR(50) NOT NULL,
    PRIMARY KEY (`user_id`, `channel_id`),
    FOREIGN KEY (`user_id`) REFERENCES `user` (`user_id`) ON UPDATE CASCADE ON DELETE CASCADE,
    FOREIGN KEY (`channel_id`) REFERENCES `channel` (`channel_id`) ON UPDATE CASCADE ON DELETE CASCADE
);
'''
```

```python
file_name = 'nddp.sql'

with open(file_name, 'w') as file:
    file.write('USE `youtube_trending_data`;')
    file.write(drop)
    file.write(country)
    file.write(user)
    file.write(channel)
    file.write(manages)
    file.write(video)
    file.write(video_stats)
    file.write(category)
    
    file.write(country_inserts)
    file.write(channel_inserts)
    file.write(video_inserts)
```
